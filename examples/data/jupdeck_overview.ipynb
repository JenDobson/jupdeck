# JupDeck: Turn Jupyter Notebooks into Presentation-Ready Slides

- Automate the creation of PowerPoint slides from Jupyter notebooks
- Save hours of copying, pasting, and formatting
- Keep your data analysis and communication in sync

# How It Works

1. Parses your notebookâ€™s structure and outputs
2. Matches content to slide layouts
3. Generates .pptx deck using python-pptx

# Automatic Slide Breaks

* New slides are automatically created from Level-1 (#) markdown headings
* Bulleted text is rendered as bullets in presentation
* Slide directives (e.g. "hide-content", "new-slide") coming soon

Paragraph text is rendered as speaker notes.  (PowerPoint and Google Slides only)

# Plot Rendering

* Sometimes there are multiple plots in a section
* The default layout puts the first plot in the lower left
