# Dataset
* The data is contained in the table presented here
* We would like the slide layout to show both the bullet points and the table

```python
# Data Exploration
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = pd.DataFrame({"x": range(10), "y": [2 * i + 1 for i in range(10)]})
data
```

# Plotting
* Now we plot the data
* The plotting code should not be rendered
* The figure should be rendered

```python
plt.figure(figsize=(6, 4))
plt.plot(data["x"], data["y"], marker="o")
plt.title("Raw Data Plot")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

# Multiple Plots
* Sometimes there are multiple plots in a section
* The default layout puts the first plot in the lower left
* All subsequent plots are overlayed lower right

```python
f1 = plt.figure()
ax1 = f1.add_subplot()
ax1.plot(np.arange(10),np.arange(10))

f2 = plt.figure()
ax2 = f2.add_subplot()
ax2.plot(np.arange(10,20,1),np.arange(10,20,1))

f3 = plt.figure()
ax3 = f3.add_subplot()
ax3.plot(np.arange(20,30,1),np.arange(20,30,1))
```

# Analysis
* To quantify the relationship between `x` and `y`, we'll use linear regression. 
* This will allow us to model `y` as a function of `x` and determine the slope and intercept that best fit the data.

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(data[["x"]], data["y"])
predicted = model.predict(data[["x"]])
plt.figure(figsize=(6, 4))
plt.scatter(data["x"], data["y"], label="Data")
plt.plot(data["x"], predicted, color="red", label="Regression Line")
plt.title("Linear Regression")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
model.coef_, model.intercept_
```

* Even though these bullet points follow the plot, they should still be rendered
* Right now the slide layout does not care whether the bullets come before or after the plot

# Add some noise
* If we add some noise to the data we will see how the R2 decreases

```python
data.y = data.y + np.random.normal(0, 1, size=len(data))

plt.plot(data.x,data.y, marker="o", linestyle="None")
plt.plot(data.x,model.predict(data[["x"]]), color="red")
```

* We can compute the R^2 on the new data...

```python
r2 = model.score(data.loc[:,["x"]],data.y)
print(f"R^2: {r2:.3f}")
```
