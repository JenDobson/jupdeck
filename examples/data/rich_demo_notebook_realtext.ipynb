# Sample Analysis Notebook
* Data analysis is a critical part of any decision-making process
* This notebook demonstrates linear regression on a simple dataset
* We demonstrate the case for noise-free and noisy data
This notebook demonstrates a basic data analysis workflow using Python. We'll begin by exploring a sample dataset, visualizing the raw data, and performing a linear regression analysis to uncover trends.

<!-- slide: new -->
# Dataset
* The data is contained in the table presented here
* We would like the slide layout to show both the bullet points and the table

```python
# Data Exploration
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = pd.DataFrame({"x": range(10), "y": [2 * i + 1 for i in range(10)]})
data
```

# Plotting
* Now we plot the data
* The plotting code should not be rendered
* The figure should be rendered

```python
plt.figure(figsize=(6, 4))
plt.plot(data["x"], data["y"], marker="o")
plt.title("Raw Data Plot")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

# Analysis
* To quantify the relationship between `x` and `y`, we'll use linear regression. 
* This will allow us to model `y` as a function of `x` and determine the slope and intercept that best fit the data.

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(data[["x"]], data["y"])
predicted = model.predict(data[["x"]])
plt.figure(figsize=(6, 4))
plt.scatter(data["x"], data["y"], label="Data")
plt.plot(data["x"], predicted, color="red", label="Regression Line")
plt.title("Linear Regression")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
model.coef_, model.intercept_
```

* Even though these bullet points follow the plot, they should still be rendered
* Right now the slide layout does not care whether the bullets come before or after the plot

# Add some noise
* If we add some noise to the data we will see how the R2 decreases

```python
data.y = data.y + np.random.normal(0, 1, size=len(data))

plt.plot(data.x,data.y, marker="o", linestyle="None")
plt.plot(data.x,model.predict(data[["x"]]), color="red")
```

* We can compute the R^2 on the new data...

```python
r2 = model.score(data.loc[:,["x"]],data.y)
print(f"R^2: {r2:.3f}")
```

# Summary and Conclusions
In this notebook, we demonstrated a basic data analysis pipeline:
- We created a synthetic dataset with a known linear relationship.
- We visualized the raw data using line plots.
- We applied linear regression to model the relationship between variables.
The results confirm a clear linear trend with a slope of approximately 2 and an intercept of 1.
